# Exercise 1.A NumPy: Manipulating Arrays

```python
import numpy as np
```

 ## Looping Through Arrays

```python
arrA = np.array([[11, 12, 13],
                 [14, 15, 16]])

print(arrA)
```

```python
for x in arrA:
    print(x * 2)
```

```python
for x in arrA:
    for y in x:
        print(y * 2)
```

## The first loop goes through each row and multiplying by 2 doubles the entire row.
## The second loop goes through each indiduval number and multiplying by 2 will print each value.

```python
for x in arrA:
    for y in x:
        loop_result = x * 2
        print(loop_result)
```

# The result is different because x represents the whole row. 

## Joining and Stacking Arrays

```python
joined_array_1 = np.concatenate((arrA, arrA), axis=0)

print(joined_array_1)
```

```python
joined_array_2 = np.concatenate((arrA, arrA), axis=1)

print(joined_array_2)
```

```python
stacked_array_1 = np.stack((joined_array_1, joined_array_1))

print(stacked_array_1)
```

```python
stacked_array_2 = np.stack((joined_array_1, joined_array_2))

print(stacked_array_2)
#  joined_array_1 and joined_array_2 have different shapes
```

```python
another_joined_array = np.concatenate((joined_array_1, arrA))

print(another_joined_array)
```

```python
another_2 = np.concatenate((joined_array_1, arrA), axis=1)

print(another_2)
# the arrays do not have the same number of rows
```

```python
h_stack = np.hstack((arrA, arrA))  # hstack stacks arrays horizontally, side by side, adding more columns.

v_stack = np.vstack((arrA, arrA))  # vstack stacks arrays vertically, top to bottom, adding more rows.

d_stack = np.dstack((arrA, arrA))  # dstack stacks arrays by depth, creating a third dimension.

print(h_stack)
print(v_stack)
print(d_stack)
```

 ## Splitting Arrays

```python
splitsky = np.array_split(joined_array_1, 2)

print(splitsky)
#  It was split into 2 separate arrays and 2-dimensional arra. 
```

```python
spliterino = np.array_split(joined_array_1, 3)

print(spliterino)
# It's 3 separate arrays,and 2-dimensional arra but uneven number of rows because only 3 groups. 
```

## Searching and Replacing Array Values

```python
result_1 = np.where(arrA % 2 == 0, arrA, 'odd')

print(result_1)
```

## Sorting and Filtering Array

```python
arrB = np.array([
    [7, 6],
    [13, 5],
    [7, 3]
])

arrB_sorted = np.sort(arrB)

print(arrB_sorted)
```

```python
print(arrB)

z = arrB < 7

arrC = arrB[z]

print(arrC)
```

```python
y = arrB >= 7

arrD = arrB[y]

print(arrD)
```
