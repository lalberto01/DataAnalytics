## Grocery Problem

```python
# Order: vegetables, fruits, snacks, beverages

bag1 = np.array([2, 3, 4, 0])
bag2 = np.array([4, 0, 1, 2])

print(bag1)
print(bag2)
```

```python
bag1_total = np.sum(bag1)
bag2_total = np.sum(bag2)

print("The first bag contains", bag1_total, "items, and the second bag contains", bag2_total, "items.")
```

```python
category_totals = np.add(bag1, bag2)

print("There are", category_totals[0], "vegetables,", category_totals[1], "fruits,", category_totals[2], "boxes of snacks, and", category_totals[3], "beverages in the two grocery bags.")
```

## Teamwork

```python
ali = np.array([5, 4, 6, 4, 2])
ben = np.array([0, 5, 5, 3, 4])
carla = np.array([3, 6, 0, 3, 4])
```

```python
all_knocks = np.array([ali, ben, carla])

print(all_knocks)
```

```python
person_totals = np.sum(all_knocks, axis=1)
group_total = np.sum(all_knocks)

print("Ali knocked on", person_totals[0], "doors.")
print("Ben knocked on", person_totals[1], "doors.")
print("Carla knocked on", person_totals[2], "doors.")
print("The group knocked on", group_total, "doors total.")
```

```python
daily_totals = np.sum(all_knocks, axis=0)
rolling_total = np.cumsum(daily_totals)
estimated_votes = rolling_total * 0.5

print(estimated_votes)
```
