## Lab 3. Employee data

This lab exercise makes updates to information extracted from
an employee database.

```python
# Employee data list of tulips 
hr_list = [
    ('0123', 'HR', 'Rebecca Yang', '69000'),
    ('0121', 'IT', 'Mark Blick', '67000'),
    ('0068', 'IT', 'Kahna Larsen', '112000'),
    ('0234', 'OPS', 'Jim Smith', '54000')
]
```

```python
# Update Mark's last name to Blick-Hawley
hr_list[1] = ('0121', 'IT', 'Mark Blick-Hawley', '67000')
```

```python
# Print updated list after Mark's name change
print(hr_list)
```

```python
# Update Jim's department to CS and salary to 60000
hr_list[3] = ('0234', 'CS', 'Jim Smith', '60000')
```

```python
# Print updated list after Mark's and Jim's changes
print(hr_list)
```

```python
# Display updated employee data on multiple lines
for employee in hr_list:
    emp_num = employee[0]
    dept = employee[1]
    name = employee[2]
    salary = int(employee[3])   # convert string to integer
    print(emp_num, "|", dept, "|", name, "|", format(salary, ","))
```
