# Exercise 2.A Using built-in functions

## Lab 2. Parsing strings – part codes

```python
def get_supplier_code(code):
    for i in range(len(code)):
        if code[i] == ":":
            return code[:i]
```

```python
def get_product_num(code):
    colon = 0
    dash = 0

    for i in range(len(code)):
        if code[i] == ":":
            colon = i
        if code[i] == "-":
            dash = i

    return code[colon+1:dash]
```

```python
def get_size(code):
    for i in range(len(code)):
        if code[i] == "-":
            return code[i+1:]
```

```python
part1 = "ACME:123-L"
part2 = "DI:12345-M"
part3 = "ACE:1-12"
```

```python
print("For", part1,
      ", the supplier is", get_supplier_code(part1) +
      ". Part number", get_product_num(part1),
      "is available in size", get_size(part1))
```

```python
print("For", part2,
      ", the supplier is", get_supplier_code(part2) +
      ". Part number", get_product_num(part2),
      "is available in size", get_size(part2))
```

```python
print("For", part3,
      ", the supplier is", get_supplier_code(part3) +
      ". Part number", get_product_num(part3),
      "is available in size", get_size(part3))
```
