# Payroll Tax Function Lab

# How to create and use custom Python functions to calculate payroll taxes. The notebook uses functions, parameters, return values, and if/else statements to calculate Social Security, Medicare, and federal tax based on gross pay and withholding codes.

```python
# SS tax function
def get_soc_sec_tax(gross_pay):
    tax = gross_pay * 0.062
    return tax
```

```python
# Medicare tax function
def get_medicare_tax(gross_pay):
    tax = gross_pay * 0.0145
    return tax
```

```python
# Function to calculate federal tax
def get_federal_tax(gross_pay, withholding_code):

    # Determine tax rate based on withholding code
    if withholding_code == 0:
        tax_rate = 0.23

    elif withholding_code == 1:
        tax_rate = 0.21

    elif withholding_code == 2:
        tax_rate = 0.195

    elif withholding_code == 3:
        tax_rate = 0.185

    else:
        tax_rate = 0.18 - ((withholding_code - 4) * 0.005)

    # Calculate federal tax
    tax = gross_pay * tax_rate

    return tax
```

```python
gross_pay = 750
withholding_code = 0

print("Person 1")
print("Social Security Tax: $", format(get_soc_sec_tax(gross_pay), ".2f"))
print("Medicare Tax: $", format(get_medicare_tax(gross_pay), ".2f"))
print("Federal Tax: $", format(get_federal_tax(gross_pay, withholding_code), ".2f"))
```
