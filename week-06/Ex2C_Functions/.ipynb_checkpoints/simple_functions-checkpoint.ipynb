# Defining custom functions

```python
# Favorite thing() function
def favorite_things(name, movie, music):
    return (
        "Name: " + name + "\n"
        "Favorite Movie: " + movie + "\n"
        "Favorite Musician/Band: " + music
    )
```

```python
print(favorite_things("Lesley", "Twilight", "Ella Langley"))
```

```python
def why_im_here(purpose):
    return purpose
```

```python
print(why_im_here("I joined the Year Up United Data Analytics program to build strong technical skills in data analysis and start a new career."))
```

```python
def favorite_place(place, why):
    return "One of my favorite places is " + place + " because " + why + "."
```

```python
print(favorite_place("New York City", "there is always something to explore and eat"))
```

```python
print(favorite_place("the gym", "it helps me relax and clear my mind"))
```

# The most efficient approach was using f-strings, because they reduce repetition and are less errors when combining variables with text. F-strings automatically format everything in a clear structure.
